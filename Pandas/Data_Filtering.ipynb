```python

#* YR_Released sütundan 2000 yılından sonra yayınlanmış filmlerin Movie_Title, Rating sütunlarını bulunan verileri getirin.
#* Ayrıca Rating sütununda bulunan verilere göre puanı en yüüksekten en aşağı olacak şekilde sıralayalım

import time

start = time.time_ns()

(
    df.query('YR_Released > 2000')
        .loc[:, ['Movie_Title', 'YR_Released', 'Rating']]
        .sort_values(by='Rating', ascending=False)
) # query filtreleme 
```

```python
df[df['YR_Released'] > 2000][['Movie_Title', 'YR_Released', 'Rating']].sort_values(by='Rating', ascending=False) # üstekinin aynısı farklı yolla
```

```python

#* Rating 7.0 dan büyük 9.0 dan küçük verileri getirelim, Raiting büyükten küçüğe doğru sıralayalım

(
    df.query('Rating > 7.0 and Rating < 9.0 ')
        .loc[:, ['Movie_Title', 'YR_Released', 'Rating']]
        .sort_values(by='Rating', ascending=False)
)
```

```python
df.columns
```

```python

#* Num_Reviews > 30000 ve YR_Released > 2000 olan filmlerin Movie_Title, Num_Reviews, YR_Released sütunlarını, Num_Reviews çoktan aza olacak şekilde listeleyelim

(
    df.query('Num_Reviews > 30000 & YR_Released > 2000 ')
        .loc[:, ['Movie_Title', 'Num_Reviews', 'YR_Released']]
        .sort_values(by='Num_Reviews', ascending=False)
)
```

```python

#* Movie_Title içerisinde "King" ifadesi geçen ve YR_Released > 2000 olan filmlerin Movie_Title, Num_Reviews, YR_Released, Num_Reviews çoktan aza olacak şekilde

(
    df.query('Movie_Title.str.contains("King", case=False) & YR_Released > 2000') # case=False yapılırsa içinde king olan kelimeleride dahil eder yazmazsak True kabul eder
        .loc[:, ['Movie_Title', 'Num_Reviews', 'YR_Released']]
        .sort_values(by='Num_Reviews', ascending=False)
)
```

```python

#* Yayın tarihi 2014 ile 2018 arasında olan filimlerin isimlerini getiriniz.

df[(df["YR_Released"] >= 2014) & (df["YR_Released"] <= 2018)][["Movie_Title", "YR_Released"]]
```

```python

#* Değerlendime sayısı (Num_Reviews) 100.000'den büyük olan yada imdb puanı 8 ile 9 arasında olan filimleir "Movie_Title", "Rating", "Num_Reviews" sütunlarını getirin.

df[(df["Num_Reviews"] >= 100000) | ((df["Rating"] >= 8) & (df["Rating"] <= 9))][["Movie_Title", "Rating", "Num_Reviews"]]
```
