# GroupBy

```python
import pandas as pd
```

```python
df = pd.read_csv(
    filepath_or_buffer='Data/imdb.csv'
)
```

```python
(
    df.groupby(by='YR_Released')
        .agg(
            {
                'Num_Reviews': 'sum'
            }
        )
        .query('Num_Reviews > 8000000 & Num_Reviews <15000000')
        .sort_values(by='Num_Reviews', ascending=False)
)
```

```python

#* Yıllara göre toplamda alınan raitingler

(
    df.groupby(by='YR_Released')
        .agg(
            {
                'Rating': 'sum'
            }
        )
        .sort_values(by='Rating', ascending=False)
)
```

```python

#* Rating göre Num_Reviews toplamı

(
    df.groupby(by='Rating')
        .agg(
            {
                'Num_Reviews': 'sum'
            }
        )
        .sort_values(by='Num_Reviews', ascending=False)
)
```
